```python
fraud_df=spark.sql('select * from fraud_datection_table t where t.fraud_flag=true')
```

```python
from pyspark.sql.functions import *
fraud_count_day_df=fraud_df.groupBy(date_format('transaction_time','dd').alias('day')).agg(count('*').alias('total_count'))
fraud_count_country_df=fraud_df.groupBy('customer_country').agg(count('*').alias('total_count'))
fraud_count_merchant_df=fraud_df.groupBy('merchant').agg(count('*').alias('total_count'))
fraud_count_customer_df=fraud_df.groupBy('user_id').agg(count('*').alias('total_count'))
fraud_count_card_df=fraud_df.groupBy('card_id').agg(count('*').alias('total_count'))
display(fraud_count_card_df)
```

```python
fraud_count_day_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storageaccountfda.dfs.core.windows.net/catalog/gold/fraud_count_day')
fraud_count_country_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storageaccountfda.dfs.core.windows.net/catalog/gold/fraud_country_count')
fraud_count_merchant_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storageaccountfda.dfs.core.windows.net/catalog/gold/fraud_merchant_count')
fraud_count_customer_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storageaccountfda.dfs.core.windows.net/catalog/gold/fraud_customer_count')
fraud_count_card_df.write.format('delta').mode('overwrite').option('mergeSchema',True).save('abfss://data@storageaccountfda.dfs.core.windows.net/catalog/gold/fraud_card_count')
```
